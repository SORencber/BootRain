**`Duration   is Float and Continous variable.`**

```
def make_discrete(time):
    if time <= 60:
        return 1
    elif (time > 60) and (time < 120):
        return 2
    else:
        return 3
movie['duration_grup'] = movie['duration'].apply(make_discrete)
display(movie.head())
movie.groupby("duration_grup")["duration"].count()
```

**3. Is it possible to convert categorical variable into numerical one? If so, how can you make that?**

Yes. For Example. we can change color column from String to integer. 
Then 1 will be Color and 0 will be Black-White in the Color Column




```
def make_int(color):
    if color=="Color":
        return 1
    else:
        return 0
movie['color'] = movie['color'].apply(make_int)
display(movie.head())
movie.color.dtype
```
