1. Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
data=df.groupby(by=['location'])['total_cases'].max()
new=data.nlargest(20)
new
```

```python
title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(10, 5))
plt.title('Total cases', fontdict = title_style)
plt.bar(new.iloc[:].index, new,color = "red")
plt.xlabel('Countries',fontdict = axis_style)
plt.ylabel('Numbers', fontdict = axis_style)
plt.xticks(rotation = 45, fontsize = 9)
plt.show()
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
Germany=df[df.location=="Germany"]
plt.figure(figsize=(20, 5))
plt.title('Histogram', fontdict = title_style)
plt.subplot(1,3,1)
plt.hist(Germany['new_deaths'], color = "brown",bins=np.arange(-10, 40, 3), alpha=.5)
plt.subplot(1,3,2)
plt.hist(Germany['new_deaths'], color = "blue",bins=np.arange(0, 30, 3), alpha=.5)
plt.subplot(1,3,3)
plt.hist(Germany['new_deaths'], color = "red",bins=np.arange(5, 30, 3), alpha=.5)
plt.show()
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

France=df[df.location=="France"]
plt.figure(figsize=(10, 6))
plt.title('New Deaths & New Cases', fontdict = title_style)
for country in ['Germany', 'France']:
    if country=="Germany":
        plt.scatter(df[df.location == country].new_deaths, df[df.location == country].new_cases,label = country,c="b")
    else:
        plt.scatter(df[df.location == country].new_deaths, df[df.location == country].new_cases, label = country,c="r",alpha=0.2)

plt.xlabel('New Deaths',fontdict = axis_style)
plt.ylabel('New Cases',fontdict = axis_style)
plt.legend(loc = 2, markerscale = 0.3)
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
plt.figure(figsize=(10, 6))
plt.title('Daily deaths', fontdict = title_style)
plt.boxplot(df['new_deaths'].dropna())
plt.xlabel('Deaths',fontdict = axis_style)
plt.ylabel('Number', fontdict = axis_style)
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart 

```python
data=pd.DataFrame()
data=df.groupby(by=['continent'])['total_cases'].max()
```

```python
plt.figure(figsize=(10, 6))
plt.title('World Total Case', fontdict = title_style)

plt.pie(data, labels=data.index, autopct='%1.1f%%',
        shadow=True, startangle=45)
plt.show()
```
