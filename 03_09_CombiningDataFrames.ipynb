````python
def file2():
    f = open("file2.csv", "w")
    f.write("id,target\n1001,0\n1002,1\n1004,0\n1005,1\n1007,0\n1008,1\n")
    f.close()

file2()
````

````python
file1 = pd.read_csv("file1.csv", index_col=0)
file2 = pd.read_csv('file2.csv', index_col = 0)
display(file1)
display(file2)
````

````python
pd.merge(file1, file2, left_on='id', right_on='id')
````

````python
pd.merge(file1, file2, left_on='id', right_on='id', how = "outer")
````

````python
pd.merge(file1, file2, left_on='id', right_on='id',  how = "left")
````

````python
pd.merge(file1, file2, left_on='id', right_on='id',  how = "right")
````
